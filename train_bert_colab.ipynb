## 3) Upload dataset folder (`data/`)

```python
from google.colab import files
import os
import zipfile

print("Upload data.zip (containing train.json, validation.json, test.json)")
uploaded_data = files.upload()

if "data.zip" not in uploaded_data:
    raise ValueError("Please upload a file named data.zip")

os.makedirs("data", exist_ok=True)
with zipfile.ZipFile("data.zip", "r") as zf:
    zf.extractall(".")

print("Dataset files:")
!ls -lh data
```
